```python
# Pelvic Incidence and sacral_slope are having higher co-relation and thus used for training and testing!
```

```python
x = data.pelvic_incidence.values.reshape(-1,1)
y = data.sacral_slope.values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=8)
```

```python
#Logistic Regression cant be used as the data is not suitable 
```

```python
linear = LinearRegression()
linear.fit(x_train,y_train)

predict_test = linear.predict(x_test)

predict_train = linear.predict(x_train)

linear_train_score = round(linear.score(x_train,y_train)*100,2)
linear_test_score = round(linear.score(x_test,y_test)*100,2)


print('train_score ' , linear_train_score )
print('test_score ' , linear_test_score)
```

```python
#R^2 (coefficient of determination) regression score function.
from sklearn.metrics import r2_score
```

```python
print("score of linear regressor:", round(r2_score(y_train,predict_train)*100)) 
#ranges between 0 and 1 where 1 is max accuracy
```
